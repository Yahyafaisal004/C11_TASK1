```python
import numpy as np
import pandas as pd
```

```python
dataframe = pd.read_csv('YFifa22.csv')
dataframe.head()
```

```python
position= input().upper()
budget = float(input())
```

```python
players= dataframe[(dataframe['player_positions'].str.contains(position)) & (dataframe['value_eur']<= budget)]
final_list= players.sort_values(by='overall',ascending=False)
if final_list.shape[0] == 0:
    print('No players available')
else:
    display(final_list[['short_name','potential','value_eur']])
```
