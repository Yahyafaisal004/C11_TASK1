```python
import numpy as np
import pandas as pd
```

```python
dataframe=pd.read_csv('fifa22.csv')
dataframe
```

Removing the long_name column

```python
dataframe.drop(columns=['long_name'],inplace=True)
dataframe.columns
```

Converting the data under short_name to lowercase

```python
dataframe['short_name']=dataframe['short_name'].str.lower()
dataframe.head(2)
```

Retrieving the 10 most valuable players

```python
valuable_players = dataframe.sort_values(by='value_eur', ascending=False).head(10)
valuable_players
```

analysis of underrated players based on potential/wage

```python
dataframe['potential/wage']= dataframe['potential']/dataframe['wage_eur']
underrated_players = dataframe.sort_values(by='potential/wage',ascending=False)
underrated_players.head(10)
```

analysis of underrated players based on value/wage

```python
dataframe['value/wage']= dataframe['value_eur']/dataframe['wage_eur']
underrated_players2 = dataframe.sort_values(by='value/wage',ascending=False)
underrated_players2.head(10)
```

We can analyze under rated players based on overall/value as well

Converting height in cm to m

```python
dataframe['height_cm']= dataframe['height_cm']/100
dataframe.rename(columns={'height_cm':'height_m'},inplace=True)
```
